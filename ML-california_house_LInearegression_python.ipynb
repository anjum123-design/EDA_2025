```python
#VIF (Variance Inflation Factor) measures how much a feature is inflated by multicollinearity 
#(i.e., how much it’s correlated with other input features).
```

** X = df.drop('MedHouseVal', axis=1)  axis=1 tells pandas to drop a column, not a row.** 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
X=df.drop('MedHouseVal', axis=1) # drop the output column
vif_list=[]
for i in range (len(X.columns)):
    vif_list.append(variance_inflation_factor(X.values,i))
vif_data=pd.DataFrame(vif_list,
                      index=X.columns,
                      columns=['VIF'])
vif_data
```

**# 4. Train-test split
This function splits your dataset into two parts:
•	One for training the model
•	One for testing (evaluating) the model
With test_size=0.2
It means:
•	80% of data → goes into training (X_train, y_train)
•	20% of data → goes into testing (X_test, y_test)
